# Ejercicio 2

Escriba un programa en Python que solicite al usuario un número entero positivo $n$ e imprima la suma de los primeros $n$ números enteros positivos.

Imprime la suma mediante como se muestra a continuación:

<center>La suma de los primeros <code>n</code> números enteros positivos es: <code>suma</code> </center>

Recuerda que la suma de los primeros $n$ números enteros positivos se puede calcular mediante la ***Fórmula de Gauss***:

$$\sum_{i=1}^{n} i = \frac{n(n + 1)}{2}$$
